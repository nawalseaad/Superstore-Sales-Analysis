### الاستنتاج

أظهرت النتائج أن فئة

**Technology**

حققت أعلى إجمالي مبيعات مقارنةً بالفئات الأخرى

كما تبين أنها ليست الأعلى في عدد الطلبات، لكنها سجلت أعلى متوسط للمبيعات، مما يشير إلى أن منتجاتها ذات قيمة بيع أعلى

```python
#ما المنتجات الأعلى مبيعًا داخل فئة
# Technology
technology_sales = df[df['Category'] == 'Technology']

technology_sales.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top_products.index, top_products.values)

plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')

plt.gca().invert_yaxis()

plt.show()
```
